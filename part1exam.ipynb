```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
order_item_df = pd.read_csv("order_item.csv")
items_df = pd.read_csv("items.csv")

merged_df = order_item_df.merge(items_df[['id', 'item_price']], left_on='item_id', right_on='id', how='left')

merged_df['total_price'] = merged_df['quantity'] * merged_df['item_price']

total_quantity = merged_df['quantity'].sum()
total_price = merged_df['total_price'].sum()
average_price_per_item = total_price / total_quantity

average_price_per_item
```

```python
customers_df = pd.read_csv("customers.csv")
items_df = pd.read_csv("Items.csv")

order_item_customer_df = order_item_df.merge(orders_df[['order_id', 'customer_id']], on='order_id', how='left')

customer_item_counts = order_item_customer_df.groupby('customer_id')['quantity'].sum()

max_customer_id = customer_item_counts.idxmax()
max_customer_quantity = customer_item_counts.max()

max_customer_details = customers_df[customers_df['id'] == max_customer_id]

max_customer_details, max_customer_quantity
```

```python
df = pd.read_csv("customers.csv")
df.dropna(subset=['gender'])
df

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='gender',palette='Spectral')
plt.title('Distribution of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
